```python
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Iris.csv')
target_names = data['Species'].unique()
data['Species'] = data['Species'].replace({'Iris-setosa': '0', 'Iris-versicolor': '1', 'Iris-virginica': '2'})
data.drop('Id',axis=1,inplace=True)
print('Dataset--->',data.head())
print('Target names--->',target_names)
```

```python
plt.rcParams["figure.figsize"] = [15,9]
import matplotlib

my_colors = ["red", "yellow", "blue"]

p = plt.scatter(x=data["SepalLengthCm"], y=data["SepalWidthCm"], c=data["Species"],cmap=matplotlib.colors.ListedColormap(my_colors))
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")

## Fix my legend
plt.legend((p,p,p), (target_names))
plt.title("Scatter Plot")
ax = plt.gca()
legend = ax.get_legend()
legend.legendHandles[0].set_color(my_colors[0])
legend.legendHandles[1].set_color(my_colors[1])
legend.legendHandles[2].set_color(my_colors[2])
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d

points = data.iloc[:,[0,1]]
#print(points.head())
vor = Voronoi(points)

# Draw boundaries
voronoi_plot_2d(vor, show_vertices=False, show_points=False)

# Add color to the points
plt.scatter(x=data["SepalLengthCm"],y=data["SepalWidthCm"],c=data["Species"],cmap=matplotlib.colors.ListedColormap(my_colors))

plt.xlabel(points.columns[0])
plt.ylabel(points.columns[1])
plt.title("Iris dataset")

# Fix legend
plt.legend((p,p,p), (target_names))
plt.title("Voronoi Diagram")
ax = plt.gca()
legend = ax.get_legend()
legend.legendHandles[0].set_color(my_colors[0])
legend.legendHandles[1].set_color(my_colors[1])
legend.legendHandles[2].set_color(my_colors[2])
plt.show()
```
